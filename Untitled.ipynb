```python
import matplotlib.pyplot as plt

import pandas as pd
```

```python
dados = pd.read_csv('./datasets/train.csv')
```

```python
dados.head()
```

```python
region = dados[dados['Country_Region']=='Brazil']
region1 = dados[dados['Country_Region']=='US']
```

```python
date = region.groupby(region['Date'], as_index=False)['TargetValue'].sum()
date1 = region1.groupby(region1['Date'], as_index=False)['TargetValue'].sum()
```

```python
date.head()
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.subplots()
ax.bar(date['Date'], date['TargetValue'], label='Brazil')
ax.scatter(date1['Date'], date1['TargetValue'], label='Brazil')
ax.legend()
plt.show
```
